```python
import os
from PIL import Image
import numpy as np
from IPython.display import Image
import torch
import PIL
import matplotlib.pyplot as plt
```

```python
#   
background = PIL.Image.open("/content/drive/MyDrive/Colab Notebooks/hci/Test.png")
plt.imshow(background)
```

```python
foreground = PIL.Image.open("/content/drive/MyDrive/Colab Notebooks/hci/Visualization4.png")
plt.imshow(foreground)
```

```python
#backgroud    
(img_h, img_w) = background.size
print(background.size)
print("img_h : " , img_h , "img_w : ", img_w)
```

```python
x = img_h
y = img_w
print("img_h : " , x, "img_w : ", y)
```

```python
# foreground  
foreground = foreground.resize((x,y))
```

```python
#  
background.paste(foreground,(0,0), foreground)
```
